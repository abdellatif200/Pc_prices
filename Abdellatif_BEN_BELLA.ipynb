```python
def rename(text):
    if text == 'Intel Core i7' or text == 'Intel Core i5' or text == 'Intel Core i3':
        return text
    else:
        if text.split()[0] == 'Intel':
            return 'Other Intel Processor'
        else:
            return 'AMD Processor'
```

```python
df['Memory'] = df['Memory'].astype(str).replace('\.0', '', regex=True)
df["Memory"] = df["Memory"].str.replace('GB', '')
df["Memory"] = df["Memory"].str.replace('TB', '000')


new = df["Memory"].str.split("+", n = 1, expand = True) #divise les valeurs de la colonne 'Memory' en deux parties en utilisant le signe '+' comme séparateur

df["first"]= new[0] # crée une nouvelle colonne 'first' qui contient la première partie de la division.
df["first"]=df["first"].str.strip() #supprime les espaces

df["second"]= new[1]

df["Layer1HDD"] = df["first"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer1SSD"] = df["first"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer1Hybrid"] = df["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer1Flash_Storage"] = df["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

df['first'] = df['first'].str.replace(r'\D', '') #supprime tous les caractères non numériques

df["second"].fillna("0", inplace = True)

df["Layer2HDD"] = df["second"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer2SSD"] = df["second"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer2Hybrid"] = df["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer2Flash_Storage"] = df["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

df['second'] = df['second'].str.replace(r'\D', '') #supprime tous les caractères non numériques

df["first"] = df["first"].astype(int)
df["second"] = df["second"].astype(int)
#Les lignes suivantes calculent les quantités totales de chaque type de stockage (HDD, SSD, Hybrid, Flash Storage) en multipliant les quantités respectives dans les colonnes 'first' et 'second' par les indicateurs binaires correspondants.
# Les résultats sont stockés dans de nouvelles colonnes 'HDD', 'SSD', 'Hybrid' et 'Flash_Storage'
df["HDD"]=(df["first"]*df["Layer1HDD"]+df["second"]*df["Layer2HDD"])
df["SSD"]=(df["first"]*df["Layer1SSD"]+df["second"]*df["Layer2SSD"])
df["Hybrid"]=(df["first"]*df["Layer1Hybrid"]+df["second"]*df["Layer2Hybrid"])
df["Flash_Storage"]=(df["first"]*df["Layer1Flash_Storage"]+df["second"]*df["Layer2Flash_Storage"])

df.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
       'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
       'Layer2Flash_Storage'],inplace=True)
```

```python
import pandas as pd

# Liste des noms de colonnes catégorielles
categorical_cols = ['Company', 'Product', 'TypeName', 'Cpu brand', 'os']
numeric_cols=['Ram','Weight','Price_euros','Touchscreen','Ips','ppi','HDD','SSD']
data = pd.get_dummies(df, columns=categorical_cols)
cols = numeric_cols + list(data.columns[data.columns.str.startswith('Company_')]) + \
       list(data.columns[data.columns.str.startswith('Product_')]) + \
       list(data.columns[data.columns.str.startswith('TypeName_')]) + \
       list(data.columns[data.columns.str.startswith('Cpu brand_')]) + \
       list(data.columns[data.columns.str.startswith('os_')])
#print(cols)
newdata = data[cols]
print(newdata)
```

```python
train_data, test_data = train_test_split(newdata, test_size=0.15, random_state=2)
```

```python
X_train = train_data.drop(['Price_euros'], axis=1)
y_train = train_data['Price_euros']
X_test = test_data.drop(['Price_euros'], axis=1)
y_test = test_data['Price_euros']
```

## Linear Regression

```python
reg = LinearRegression().fit(X_train, y_train)
y_pred = reg.predict(X_test)
r2 = r2_score(y_test, y_pred)
print("Coefficient de détermination (R²) : ", r2)
```

## lasso

```python
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)

# Faire des prédictions sur les données de test
y_pred = lasso.predict(X_test)

# Évaluer la performance du modèle
r2 = r2_score(y_test, y_pred)
print("Coefficient de détermination (R²) : ", r2)
```

## KNeighbors

```python
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)

# Faire des prédictions sur les données de test
y_pred = knn.predict(X_test)

# Évaluer la performance du modèle
r2 = r2_score(y_test, y_pred)
print("Coefficient de détermination (R²) : ", r2)
```

## DecisionTree

```python
# Créer et ajuster le modèle d'arbre de décision
tree = DecisionTreeRegressor(max_depth=3)
tree.fit(X_train, y_train)

# Faire des prédictions sur les données de test
y_pred = tree.predict(X_test)

# Évaluer la performance du modèle
r2 = r2_score(y_test, y_pred)
print("Coefficient de détermination (R²) : ", r2)
```

## SVM

```python
from sklearn.svm import SVR

# Créer un modèle SVR et l'entraîner
svr = SVR(kernel='rbf', C=1e3, gamma=0.1)
svr.fit(X_train, y_train)


y_pred = svr.predict(X_test)
r2 = r2_score(y_test, y_pred)
print("Précision du modèle : ", r2)
```

## Random Forest

```python
# Créer et ajuster le modèle de forêt aléatoire
rf = RandomForestRegressor(n_estimators=100, max_depth=3)
rf.fit(X_train, y_train)

# Faire des prédictions sur les données de test
y_pred = rf.predict(X_test)

# Évaluer la performance du modèle
r2 = r2_score(y_test, y_pred)
print("Coefficient de détermination (R²) : ", r2)
```

## AdaBoost

```python

# Créer et ajuster le modèle AdaBoost
adaboost = AdaBoostRegressor(n_estimators=100, learning_rate=0.1)
adaboost.fit(X_train, y_train)

# Faire des prédictions sur les données de test
y_pred = adaboost.predict(X_test)

# Évaluer la performance du modèle
r2 = r2_score(y_test, y_pred)
print("Coefficient de détermination (R²) : ", r2)
```

## Gradient Boosting

```python
# Créer et ajuster le modèle Gradient Boosting
gradient_boosting = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1)
gradient_boosting.fit(X_train, y_train)

# Faire des prédictions sur les données de test
y_pred = gradient_boosting.predict(X_test)

# Évaluer la performance du modèle
r2 = r2_score(y_test, y_pred)
print("Coefficient de détermination (R²) : ", r2)
```
